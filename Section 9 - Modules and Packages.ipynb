# __ name __ and "__ main __"

In Python, __ name __ is a special variable that gets automatically set depending on how the code is being run.

When you run a Python script directly (e.g., python myscript.py), the value of __ name __ is set to "__ main __". This means the script is being executed as the main program.

When you import a Python file as a module in another script, __ name __ is set to the module's name, not "__ main __". This allows code to behave differently when it's imported rather than run directly.

Common Use:
You often see this block in Python scripts:

```python
if __name__ == "__main__":
    # This code runs only if the file is executed directly
    print("Running as a standalone script!")
```

If run directly: The condition is True, and the code inside the if block runs.
If imported: The condition is False, so the block is skipped.
This structure is useful for writing code that can both be run directly or imported without running everything.
